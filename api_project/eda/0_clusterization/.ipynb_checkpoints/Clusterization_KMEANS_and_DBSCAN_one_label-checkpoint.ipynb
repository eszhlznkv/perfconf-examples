## Импортируем все необходимое

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, KMeans

# Загрузка данных из CSV файлов
data = pd.read_csv('/Users/evgenii/Documents/GitHub/perfconf/perfcong-jmeter-examples/data_api/ar_api.jtl')
data_new_release = pd.read_csv('/Users/evgenii/Documents/GitHub/perfconf/perfcong-jmeter-examples/data_api/ar_api_power.jtl')
```

```python
# Получение уникальных значений label для обоих наборов данных
sample_labels = pd.concat([data['label'], data_new_release['label']]).unique()
```

```python
sample_labels = np.delete(sample_labels, 0)
```

```python
sample_labels
```

```python
# Определение количества графиков
n_labels = len(sample_labels)
fig, axes = plt.subplots(n_labels, 4, figsize=(15, 6 * n_labels))

# Цикл по каждой метке для выполнения кластеризации
for i, label in enumerate(sample_labels):
    # Фильтрация данных по текущей метке
    data_filtered = data[data['label'] == label]
    data_new_release_filtered = data_new_release[data_new_release['label'] == label]

    # Если данных для метки нет, пропускаем итерацию
    if data_filtered.empty or data_new_release_filtered.empty:
        continue
    
    # Выбор признака для кластеризации
    elapsed_data = data_filtered[['elapsed']]
    elapsed_data_new_release = data_new_release_filtered[['elapsed']]

    # Нормализация данных
    scaler = StandardScaler()
    data_normalized = scaler.fit_transform(elapsed_data)
    data_new_release_normalized = scaler.fit_transform(elapsed_data_new_release)
    
    # Определение параметров DBSCAN (eps и min_samples) для обоих наборов данных
    eps = 0.5
    min_samples = 3
    
    # Применение DBSCAN и K-Means для первого набора данных
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan_labels = dbscan.fit_predict(data_normalized)

    kmeans = KMeans(n_clusters=2, random_state=42)
    kmeans_labels = kmeans.fit_predict(data_normalized)
    
    # Применение DBSCAN и K-Means для второго набора данных
    dbscan_new = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan_labels_new = dbscan_new.fit_predict(data_new_release_normalized)

    kmeans_new = KMeans(n_clusters=2, random_state=42)
    kmeans_labels_new = kmeans_new.fit_predict(data_new_release_normalized)
    
    # Построение графиков для первого набора данных
    sns.scatterplot(x=elapsed_data['elapsed'], y=dbscan_labels, hue=dbscan_labels, ax=axes[i, 0], palette='viridis', legend=None)
    axes[i, 0].set_title(f'DBSCAN for {label} - First Dataset')
    axes[i, 0].set_xlabel('Elapsed Time (ms)')
    axes[i, 0].set_ylabel('Cluster')

    sns.scatterplot(x=elapsed_data['elapsed'], y=kmeans_labels, hue=kmeans_labels, ax=axes[i, 1], palette='viridis', legend=None)
    axes[i, 1].set_title(f'K-Means for {label} - First Dataset')
    axes[i, 1].set_xlabel('Elapsed Time (ms)')
    axes[i, 1].set_ylabel('Cluster')

    # Overlay K-Means centroids for First Dataset
    centroids = scaler.inverse_transform(kmeans.cluster_centers_)
    axes[i, 1].scatter(centroids[:, 0], [0] * len(centroids), s=300, c='red', label='Centroids', marker='X')

    # Построение графиков для второго набора данных
    sns.scatterplot(x=elapsed_data_new_release['elapsed'], y=dbscan_labels_new, hue=dbscan_labels_new, ax=axes[i, 2], palette='viridis', legend=None)
    axes[i, 2].set_title(f'DBSCAN for {label} - Second Dataset')
    axes[i, 2].set_xlabel('Elapsed Time (ms)')
    axes[i, 2].set_ylabel('Cluster')

    sns.scatterplot(x=elapsed_data_new_release['elapsed'], y=kmeans_labels_new, hue=kmeans_labels_new, ax=axes[i, 3], palette='viridis', legend=None)
    axes[i, 3].set_title(f'K-Means for {label} - Second Dataset')
    axes[i, 3].set_xlabel('Elapsed Time (ms)')
    axes[i, 3].set_ylabel('Cluster')

    # Overlay K-Means centroids for Second Dataset
    centroids_new = scaler.inverse_transform(kmeans_new.cluster_centers_)
    axes[i, 3].scatter(centroids_new[:, 0], [0] * len(centroids_new), s=300, c='red', label='Centroids', marker='X')

plt.tight_layout()
plt.show()
```
